```python
import fiftyone as fo
from PIL import Image
from torchvision.transforms import functional as func
from mmdet.apis import inference_detector, show_result_pyplot
from PIL import Image
from torchvision.transforms import functional as func
import fiftyone as fo
import pandas as pd
import numpy as np
from pycocotools.coco import COCO
```

```python
dataset = fo.Dataset.from_images_dir(images_dir='/opt/ml/detection/dataset/test')
classes = ("General trash", "Paper", "Paper pack", "Metal", "Glass", 
           "Plastic", "Styrofoam", "Plastic bag", "Battery", "Clothing")
session = fo.launch_app(dataset) ## unlabeled dataset visualization
```

```python
file = '/opt/ml/detection/mmdetection/boostcamp/submissions/vfnet_res2netCenternet_wbf.csv'
submission_df =pd.read_csv(file)

image_ids = submission_df['image_id'].tolist()

annotation = '/opt/ml/detection/dataset/test.json'
coco = COCO(annotation)

output=[[] for _ in range(len(image_ids))]

for i, image_id in enumerate(image_ids):
    prediction_string = ''
    box_list = []
    scores_list = []
    labels_list = []
    image_info = coco.loadImgs(i)[0]
    
    predict_string = submission_df[submission_df['image_id'] == image_id]['PredictionString'].tolist()[0]
    predict_list = str(predict_string).split()
            
    predict_list = np.reshape(predict_list, (-1, 6))

    for box in predict_list[:, 2:6].tolist():
        box[0] = float(box[0])# / image_info['width']
        box[1] = float(box[1])# / image_info['height']
        box[2] = float(box[2])# / image_info['width']
        box[3] = float(box[3])# / image_info['height']
        box_list.append(box)

    scores_list =list(map(float, predict_list[:, 1].tolist()))
    labels_list =list(map(int,map(float, predict_list[:, 0].tolist())))
    values_list = list(zip(box_list,scores_list))
    output[i] = list(zip(labels_list,scores_list,box_list))
```

```python
predictions_view = dataset.limit(len(dataset))

# Add predictions to samples
with fo.ProgressBar() as pb:
    for idx,sample in enumerate(pb(predictions_view)):
        c, h, w = 3,1024,1024
        # Convert detections to FiftyOne format
        detections = []
        for label, score, box in output[idx]:
            # Convert to [top-left-x, top-left-y, width, height]
            # in relative coordinates in [0, 1] x [0, 1]
            x1, y1, x2, y2 = box
            rel_box = [x1 / w, y1 / h, (x2 - x1) / w, (y2 - y1) / h]

            detections.append(
                fo.Detection(
                    label=classes[label],
                    bounding_box=rel_box,
                    confidence=score
                )
            )

        # Save predictions to dataset
        sample["predictions"] = fo.Detections(detections=detections)
        sample.save()
```

```python
session.view = predictions_view
session.freeze()
```
