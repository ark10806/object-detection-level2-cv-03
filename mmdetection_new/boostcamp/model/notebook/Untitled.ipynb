```python
import torch
import cv2
import numpy as np
import albumentations
import matplotlib.pyplot as plt
```

```python
!ls ../../../dataset/train
```

```python
image = cv2.imread('../../../dataset/train/2468.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB).astype(np.float32)
image /= 255.0
```

```python
A = albumentations.augmentations.geometric.transforms.PiecewiseAffine((0.08,0.08),nb_rows=4,nb_cols=4,p=1)
%time transposed_image = A(image=image)
plt.imshow(transposed_image['image'])
```

```python
image *=255
image = image.astype(np.uint8)
```

```python
A = albumentations.augmentations.transforms.CLAHE(p=1,always_apply=True,tile_grid_size=(36,36))
%time transposed_image = A(image=image.astype(np.uint8))
plt.imshow(transposed_image['image'])
```

```python
A = albumentations.Affine(p=1.,shear=[-30,30],scale=[0.5,1.5])
%time transposed_image = A(image=image)
plt.imshow(transposed_image['image'])
```
